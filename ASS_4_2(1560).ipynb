#Task Description-2
• One-shot: Provide one example: Input: [1, 2, 3, 4], Output: 10 to help AI generate a
function that calculates the sum of elements in a list.
Expected Output-2
• A correct conversion function guided by the single example.

PROMPT

Input: [1, 2, 3, 4]  
Output: 10  

Based on this example, write a Python function that takes a list of numbers as input and returns the sum of all its elements.


```python
def calculate_sum(input_list):
    """
    Calculates the sum of all elements in a given list.

    Args:
        input_list (list): The list of numbers to be summed.

    Returns:
        int or float: The sum of the elements in the list.
    """
    return sum(input_list)

print("The 'calculate_sum' function has been defined.")
```

### Explanation of `calculate_sum` Function

The `calculate_sum` function is straightforward and efficiently computes the sum of all numeric elements within a list. Here's how it works:

1.  **Function Definition**: The function `calculate_sum(input_list)` takes a single argument, `input_list`, which is expected to be a list containing numbers (integers or floats).

2.  **Using `sum()`**: The core of the function is the built-in Python function `sum()`. This function is designed to take an iterable (like a list) of numbers and return their sum.

3.  **Return Value**: The function directly returns the result of `sum(input_list)`, which is the total sum of all numbers in the provided list.

```python
# Sample Input
sample_input = [1, 2, 3, 4]

# Calculate the sum using the function
sample_output = calculate_sum(sample_input)

# Display Input and Output
print(f"Input: {sample_input}")
print(f"Output: {sample_output}")
```

#Task Description-3

• Few-shot: Give 2–3 examples to create a function that extracts digits from an
alphanumeric string.

Expected Output-3

• Accurate function that returns only the digits from alphanumeric string.

Few-Shot Prompt: Write a Python function called extract_digits that takes an alphanumeric string as input and returns a string containing only the digits from the input string. The function should be robust and handle various input formats.

```python
def extract_digits(alphanumeric_string):
    """
    Extracts digits from an alphanumeric string.

    Args:
        alphanumeric_string (str): The input string containing alphanumeric characters.

    Returns:
        str: A string containing only the digits from the input string, in their original order.
    """
    digits = [char for char in alphanumeric_string if char.isdigit()]
    return "".join(digits)

print("The 'extract_digits' function has been defined.")
```

This code extracts only the digits from a string, removing everything else.

The function uses a regular expression (re.sub) to remove anything that's not a number (0-9).

It returns a string with just the digits from the input.

Test cases show how it works with different types of input strings.

```python
print("### Sample Inputs and Outputs for `extract_digits` function\n")

sample_inputs = [
    "a1b2c3",
    "x9y8z7",
    "abc123xyz"
]

for input_str in sample_inputs:
    output_str = extract_digits(input_str)
    print(f"Input: \"{input_str}\", Output: \"{output_str}\"")
```

## TASK 4: ZERO-SHOT PROMPTING

### Prompt Used (Zero-shot)

*"Write a Python function to count the number of vowels in a string."*

Zero-Shot Prompt: Write a Python function called count_vowels that takes a string as input and returns the total number of vowels (a, e, i, o, u, case-insensitive) present in the string.

```python
def count_vowels_few_shot(text):
    """
    Counts the number of vowels (a, e, i, o, u, case-insensitive) in a given string, guided by examples.

    Args:
        text (str): The input string.

    Returns:
        int: The number of vowels in the string.
    """
    vowels = "aeiou"
    count = 0
    for char in text:
        if char.lower() in vowels:
            count += 1
    return count

print("The 'count_vowels_few_shot' function has been defined.")
```

```python
print("### Sample Inputs and Outputs for `count_vowels_few_shot` function\n")

sample_tests = [
    "hello",
    "Education",
    "Rhythm",
    "Programming is fun"
]

for text_input in sample_tests:
    vowel_count = count_vowels_few_shot(text_input)
    print(f"Input: \"{text_input}\", Vowel Count: {vowel_count}")
```

## Comparison of Zero-shot and Few-shot Vowel Counter Outputs

Zero-shot prompting provides only the instruction, so the AI relies completely on its general knowledge to generate the solution.

One-shot prompting includes one example, which helps the AI understand the expected input–output pattern.

Zero-shot works well for simple and common problems but may lead to ambiguity in complex tasks.

One-shot prompting usually gives more accurate and aligned outputs because the example guides the model’s logic.

#Task Description-5
• Use few-shot prompting with 3 sample inputs to generate a function that determines

the minimum of three numbers without using the built-in min() function.

Expected Output-5

• A function that handles all cases with correct logic based on example patterns.

Few-Shot Prompt: Write a Python function called find_minimum_of_three that takes three numbers (num1, num2, num3) as input and determines the smallest among them. This function should not use Python's built-in min() function. Instead, it should use conditional logic to compare the numbers.

Examples:

Input: 3, 7, 5, Output: 3

Input: 10, 2, 8, Output: 2

Input: 4, 4, 6, Output: 4

```python
def find_minimum_of_three(a, b, c):
    """
    Determines the minimum of three numbers without using the built-in min() function.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.
        c (int or float): The third number.

    Returns:
        int or float: The minimum of the three numbers.
    """
    minimum = a
    if b < minimum:
        minimum = b
    if c < minimum:
        minimum = c
    return minimum

print("The 'find_minimum_of_three' function has been defined.")
```

### Explanation of `find_minimum_of_three` Function

This function finds the smallest of three given numbers without using Python's built-in `min()` function. Here's how it works:

1.  **Function Definition**: The function `find_minimum_of_three(a, b, c)` takes three arguments: `a`, `b`, and `c`, which are the numbers to be compared.

2.  **Initialization**: It starts by assuming that the first number, `a`, is the `minimum`.

3.  **Comparison with the Second Number**: It then checks if `b` is smaller than the current `minimum`. If `b` is indeed smaller, `minimum` is updated to `b`.

4.  **Comparison with the Third Number**: Next, it checks if `c` is smaller than the current `minimum`. If `c` is smaller, `minimum` is updated to `c`.

5.  **Return Value**: After comparing all three numbers, the function returns the final `minimum` value found.

```python
print("### Sample Inputs and Outputs for `find_minimum_of_three` function\n")

sample_inputs_min = [
    (3, 7, 5),
    (10, 2, 8),
    (4, 4, 6)
]

for a, b, c in sample_inputs_min:
    result = find_minimum_of_three(a, b, c)
    print(f"Input: {a}, {b}, {c}, Output: {result}")
```
