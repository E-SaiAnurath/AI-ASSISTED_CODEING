```python
#naive approach
import math
import time

def is_prime_naive(n):
    """
    Checks if a number is prime using a naive approach.
    A number is prime if it is greater than 1 and has no divisors
    other than 1 and itself. We check divisibility from 2 up to n-1.
    """
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print("Naive Prime Checker:")
num = int(input("Enter a number to check: "))
print(f"Is {num} prime? {is_prime_naive(num)}")
```

```python
#optimized approach
def is_prime_optimized(n):
    """
    Checks if a number is prime using an optimized approach.
    We only need to check for divisors up to the square root of n.
    Also, we handle even numbers separately and then check only odd divisors.
    """
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    # Check divisors from 5 up to sqrt(n)
    # We can skip numbers that are multiples of 2 or 3
    # All primes greater than 3 are of the form 6k ± 1
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
print("Optimized Prime Checker:")
num = int(input("Enter a number to check: "))
print(f"Is {num} prime? {is_prime_optimized(num)}")
```

### Explanation of Time Complexity and Performance Comparison

#### 1. Naive Approach (`is_prime_naive`)

**Algorithm:** This method checks every integer from 2 up to `n-1` to see if it divides `n`. If any number divides `n` evenly, then `n` is not prime.

**Time Complexity: O(n)**

*   In the worst case (when `n` is a prime number), the loop will iterate approximately `n-2` times. For example, to check if 101 is prime, it might perform almost 99 divisions.
*   As `n` grows, the number of operations increases linearly with `n`.

**Example Performance:** For a large prime number like 1,000,003, this function would perform close to 1 million division operations.

#### 2. Optimized Approach (`is_prime_optimized`)

**Algorithm:** This method incorporates several optimizations:

1.  **Checking up to `sqrt(n)`:** If a number `n` has a divisor `d > sqrt(n)`, then it must also have a divisor `n/d < sqrt(n)`. Therefore, we only need to check for divisors up to `sqrt(n)`. This significantly reduces the range of numbers to check.
2.  **Handling 2 and 3:** It explicitly checks for divisibility by 2 and 3. This allows us to start checking from 5 and then increment by 6, as all primes greater than 3 are of the form `6k ± 1` (e.g., 5, 7, 11, 13, 17, 19...). This pattern skips many composite numbers.

**Time Complexity: O(sqrt(n))**

*   In the worst case, the loop iterates approximately `sqrt(n)/6` times (due to checking `i` and `i+2` and incrementing `i` by 6).
*   As `n` grows, the number of operations increases much slower than linearly.

**Example Performance:** For a large prime number like 1,000,003:
    *   `sqrt(1,000,003)` is approximately `1000`.
    *   The optimized function would perform only a few hundred division operations, a drastic reduction compared to the naive approach.

#### Performance Comparison

Let's compare the number of operations for a moderately large number:

*   **For `n = 1,000,000`:**
    *   Naive: Approximately 1,000,000 operations.
    *   Optimized: Approximately `sqrt(1,000,000) = 1,000` operations (or even less due to the 6k ± 1 optimization). This is a 1000-fold improvement!

The optimized approach provides a substantial performance improvement, especially for larger numbers, because it drastically reduces the number of division checks required. While the naive approach grows linearly with `n`, the optimized approach grows with the square root of `n`, making it much more efficient for real-world applications.

```python
# Let's test the performance difference with a larger number

large_number = 10_000_003 # A large prime number

print(f"Testing prime checking for {large_number}:")

start_time = time.time()
is_prime_naive(large_number)
end_time = time.time()
print(f"Naive approach took: {end_time - start_time:.6f} seconds")

start_time = time.time()
is_prime_optimized(large_number)
end_time = time.time()
print(f"Optimized approach took: {end_time - start_time:.6f} seconds")
```

```python
def fibonacci_recursive(n):
    """
    Calculates the nth Fibonacci number using a recursive approach.
    """
    if n <= 0:
        return 0       # Base case: F(0) = 0
    elif n == 1:
        return 1       # Base case: F(1) = 1
    else:
        # Recursive case: F(n) = F(n-1) + F(n-2)
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

print("Recursive Fibonacci Function:")

# 🔥 Multiple inputs in one line
numbers = list(map(int, input("Enter multiple numbers separated by space: ").split()))

for num in numbers:
    print(f"Fibonacci({num}) = {fibonacci_recursive(num)}")
```

### Explanation of Recursive Fibonacci

The `fibonacci_recursive(n)` function calculates the nth Fibonacci number by directly implementing the mathematical definition of the Fibonacci sequence, where each number is the sum of the two preceding ones.

**How Recursion Works Here:**

1.  **Problem Decomposition:** To find `F(n)`, the function breaks it down into finding `F(n-1)` and `F(n-2)`. Each of these subproblems is handled by another call to the `fibonacci_recursive` function.
2.  **Base Cases:**
    *   `if n <= 0: return 0`:
        This is the first base case. If the input `n` is 0 or less, the function immediately returns 0. This is essential because it provides a definite stopping point for the recursion, preventing an infinite loop of calls for non-positive numbers.
    *   `elif n == 1: return 1`:
        This is the second base case. If `n` is 1, the function returns 1. This is another fundamental definition of the Fibonacci sequence and acts as a stopping point for the recursive calls that eventually lead to `n=1`.
3.  **Recursive Calls:**
    *   `else: return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)`:
        This is the heart of the recursion. If `n` is greater than 1, the function calls itself twice: once for `n-1` and once for `n-2`. These calls generate new subproblems that will continue to decompose until they hit one of the base cases. The results from these two recursive calls are then added together to produce the final `F(n)`.



This demonstrates how the function unwinds by making smaller calls until it reaches a known value (the base cases), and then builds the solution back up.

## Security Flaws in the Initial Login System

The generated login system, as intentionally designed, contains several critical security flaws. These flaws are fundamental and would make a real-world application highly vulnerable.

### 1. Plain-Text Password Storage
*   **Problem:** The `create_user` function stores passwords directly as received (`users[username] = password`) in a dictionary (`users`). In a real application, this would translate to storing passwords in plain text in a database.
*   **Security Risk:** If the `users` dictionary (or database) is ever compromised, all user passwords are immediately exposed to an attacker. This is extremely dangerous as users often reuse passwords across multiple services.

### 2. Insecure Password Comparison
*   **Problem:** The `login_user` function verifies a user's identity by directly comparing the provided password with the plain-text password stored (`if users[username] == password`).
*   **Security Risk:** This practice requires the system to always know and store the user's actual password. Modern security practices dictate that a system should never know a user's actual password, only a one-way hashed representation of it.

### 3. Lack of Robust Input Validation (Password Complexity)
*   **Problem:** The system accepts any string as a password without enforcing complexity rules (e.g., minimum length, requirement for uppercase, lowercase, numbers, or special characters).
*   **Security Risk:** This allows users to set very weak and easily guessable passwords (like "password123" or "secret456" used in the example). Weak passwords are susceptible to brute-force attacks or dictionary attacks, making accounts easy to compromise.

### 4. No Rate Limiting on Login Attempts
*   **Problem:** The system doesn't restrict the number of login attempts for a given username or IP address.
*   **Security Risk:** This absence makes the system vulnerable to brute-force attacks, where an attacker can try an unlimited number of password combinations until they find the correct one.

```python
import hashlib

def hash_password(password):
    """
    Hashes a password using SHA-256 for secure storage.
    """
    return hashlib.sha256(password.encode()).hexdigest()

def validate_password(password):
    """
    Validates password complexity (example: min 8 chars, 1 uppercase, 1 lowercase, 1 digit).
    """
    if len(password) < 8:
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isdigit() for char in password):
        return False
    return True

def create_user_secure(username, password, users):
    """
    Creates a new user with hashed password and input validation.
    """
    if not validate_password(password):
        print("Error: Password does not meet complexity requirements (min 8 chars, 1 uppercase, 1 lowercase, 1 digit).")
        return False
    if username in users:
        print("Error: Username already exists.")
        return False

    hashed_password = hash_password(password)
    users[username] = hashed_password
    print(f"User '{username}' created successfully with a hashed password.")
    return True

def login_user_secure(username, password, users):
    """
    Attempts to log in a user by comparing the hash of the provided password
    with the stored hashed password.
    """
    if username not in users:
        print("Error: Username not found.")
        return False

    hashed_provided_password = hash_password(password)
    if users[username] == hashed_provided_password:
        print(f"User '{username}' logged in successfully.")
        return True
    else:
        print("Error: Incorrect password.")
        return False

# --- Main program flow with secure functions ---
if __name__ == "__main__":
    users_db_secure = {}

    print("\n--- Revised Secure Login System ---")

    while True:
        print("\nOptions:")
        print("1. Create User (Secure)")
        print("2. Login (Secure)")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            username = input("Enter new username: ")
            password = input("Enter new password: ")
            create_user_secure(username, password, users_db_secure)
        elif choice == '2':
            username = input("Enter username: ")
            password = input("Enter password: ")
            login_user_secure(username, password, users_db_secure)
        elif choice == '3':
            print("Exiting secure login system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```

## Revised Login System: Addressing Security Flaws

The revised login system incorporates several key improvements to enhance security, directly addressing the flaws identified in the initial version.

### 1. Password Hashing for Secure Storage
*   **Improvement:** The `hash_password` function uses `hashlib.sha256` to create a one-way hash of the password before storage. The `create_user_secure` function now stores this hash instead of the plain-text password (`users[username] = hashed_password`). During login, `login_user_secure` hashes the provided password and compares this hash to the stored hash.
*   **Security Benefit:** The system no longer stores or directly handles plain-text passwords. Even if the `users_db_secure` (or a database) is compromised, an attacker would only get the password hashes, not the actual passwords. Reversing a strong hash to get the original password is computationally infeasible.

### 2. Input Validation (Password Complexity)
*   **Improvement:** The `validate_password` function enforces complexity rules (minimum length of 8 characters, at least one uppercase letter, one lowercase letter, and one digit). The `create_user_secure` function calls this validator, and if the password doesn't meet the criteria, user creation is rejected.
*   **Security Benefit:** This helps users create stronger passwords, making them less susceptible to brute-force and dictionary attacks. This significantly raises the bar for attackers trying to guess passwords.

### 3. Separation of Concerns for Clarity
*   **Improvement:** Dedicated functions like `hash_password` and `validate_password` are created for specific security-related tasks, making the code more modular and easier to understand, maintain, and audit for security.

### Best Practices for Secure Authentication (Note):
While the revised code significantly improves security, a real-world production-ready authentication system would require even more robust measures, including:

### Secure Authentication Best Practices

Secure user authentication is paramount for protecting user data and maintaining the integrity of any system. The initial login system demonstrated common vulnerabilities; here's how to mitigate them and best practices for robust authentication:

#### 1. Password Hashing (Crucial for Security)
*   **What it is:** Instead of storing passwords in plain text, a one-way cryptographic function (a hash function) is used to transform the password into a fixed-size string of characters. This hash is stored, not the original password.
*   **Why it's important:** If a database is compromised, attackers will only get hashes, not the actual passwords. Good hash functions are designed to be irreversible, meaning it's computationally infeasible to derive the original password from its hash.
*   **Mitigates:** Plain-text password storage vulnerability.

#### 2. Salting (Enhances Hashing Security)
*   **What it is:** A unique, random string (the \"salt\") is added to each password *before* it's hashed. The salt is then stored along with the hash.
*   **Why it's important:** Salting prevents \"rainbow table\" attacks (precomputed tables of common password hashes) and ensures that identical passwords have different hashes, making it harder for attackers to identify and exploit common passwords or crack multiple accounts if one is compromised.
*   **Mitigates:** Plain-text password storage and makes brute-force attacks against multiple accounts less efficient.

#### 3. Secure Password Comparison
*   **What it is:** When a user attempts to log in, the system retrieves the stored salt and hash for that username. It then salts the *provided* password with the stored salt and hashes it. Finally, it compares this newly generated hash with the stored hash. If they match, the password is correct.
*   **Why it's important:** The original password is never needed or exposed during the comparison process.
*   **Mitigates:** Insecure password comparison vulnerability.

#### 4. Robust Input Validation (Password Complexity and Other Inputs)
*   **What it is:** Enforcing rules for password creation (e.g., minimum length, requiring a mix of uppercase, lowercase, numbers, and special characters). It also applies to other inputs to prevent injection attacks.
*   **Why it's important:** Strong passwords are much harder to guess or crack via brute-force or dictionary attacks. For other inputs, validation prevents malicious code or data from entering the system.
*   **Mitigates:** Weak password vulnerability and various injection attacks.

#### 5. Rate Limiting on Login Attempts
*   **What it is:** Restricting the number of login attempts from a particular user, IP address, or device within a given time frame.
*   **Why it's important:** This prevents or significantly slows down brute-force attacks, where an attacker repeatedly tries different password combinations.
*   **Mitigates:** Brute-force attacks against login credentials.

#### 6. Two-Factor Authentication (2FA) / Multi-Factor Authentication (MFA)
*   **What it is:** Requiring users to provide two or more distinct forms of identification before granting access (e.g., something they know like a password, something they have like a phone, or something they are like a fingerprint).
*   **Why it's important:** Even if an attacker compromises a password, they still cannot access the account without the second factor. This is a critical layer of defense against credential theft. This is a critical layer of defense against credential theft.
*   **Mitigates:** Credential stuffing and password reuse attacks.

*   **Regular Security Audits:** Continuously monitor and audit the authentication system for new vulnerabilities and misconfigurations.
*   **Dependency Management:** Keep all libraries and frameworks up-to-date to patch known security vulnerabilities.

By adopting these best practices, authentication systems can provide a much higher level of security, safeguarding user accounts and data effectively.

### 2. Lack of Anonymization or Masking
*   **Problem:** No attempt is made to mask, hash, or anonymize any part of the logged data. The information is stored in its raw form.
*   **Privacy Risk:** Without anonymization or masking, the logs become a rich target for attackers. For instance, an IP address can reveal a user's general geographical location, and a username is often a direct identifier. Masking parts of an IP address (e.g., `192.168.1.xxx`) or hashing usernames could reduce this risk.

### 3. Indefinite Retention (Implied)
*   **Problem:** The script doesn't define any log rotation or retention policies. Logs could potentially accumulate indefinitely.
*   **Privacy Risk:** The longer sensitive data is stored, the higher the risk of it being compromised or misused. Indefinite retention increases the 'blast radius' of a data breach and can lead to non-compliance with data protection regulations (like GDPR or CCPA).

### 5. Lack of Context for Data Minimization
*   **Problem:** The script logs the `username` and `ip_address` for every activity without clear justification for why both pieces of information are necessary for *all* logged events. Privacy-aware logging often advocates for logging the *minimum* necessary data.

By addressing these points, the revised logging script will aim to be more privacy-preserving while still serving its purpose of monitoring user activity.

```python
import logging
import datetime
import hashlib # For hashing sensitive data

# 1. Configure logging for the improved script
# We'll use a new log file name to distinguish from the initial one
logging.basicConfig(
    filename='user_activity_secure.log',  # Log to this new file
    level=logging.INFO,                   # Log INFO messages and above
    format='%(asctime)s - %(levelname)s - %(message)s' # Define the log format
)

# 2. Helper function to anonymize IP addresses
def anonymize_ip(ip_address):
    """
    Anonymizes the last octet of an IPv4 address to protect user privacy.
    For IPv6, it might mask a larger part or hash it entirely.
    """
    if not ip_address:
        return "[N/A]"
    parts = ip_address.split('.')
    if len(parts) == 4:
        # Mask the last octet for IPv4
        return '.'.join(parts[:3]) + '.XXX'
    elif ':' in ip_address:
        # Simple masking for IPv6 (e.g., hash or mask more parts)
        return "IPv6_MASKED" # Or implement more sophisticated IPv6 anonymization
    return ip_address # Return original if not recognized IPv4/IPv6 format

# 3. Helper function to hash sensitive identifiers like username
def hash_identifier(identifier):
    """
    Hashes a string identifier using SHA-256. This is a one-way operation.
    """
    if not identifier:
        return "[ANONYMOUS]"
    return hashlib.sha256(identifier.encode()).hexdigest()

# 4. Define the improved logging function
def log_activity_secure(username, ip_address, event_description, minimal_logging=True):
    """
    Logs user activity with privacy-preserving techniques.
    - Anonymizes IP addresses.
    - Hashes usernames if minimal_logging is True.
    - Logs a timestamp and a general event description.
    """
    current_time = datetime.datetime.now()

    # Anonymize IP address
    anonymized_ip = anonymize_ip(ip_address)

    # Decide whether to hash or keep username based on minimal_logging
    if minimal_logging:
        logged_username = hash_identifier(username)
    else:
        logged_username = username # For cases where full username might be justified (e.g., admin actions)

    log_message = (
        f"Timestamp: {current_time}, "
        f"User: {logged_username}, "
        f"IP: {anonymized_ip}, "
        f"Event: {event_description}"
    )
    logging.info(log_message)

# 5. Simulate user activities using the secure logging function
print("Simulating user activities with privacy-preserving logging. Check 'user_activity_secure.log' for output.")

# Example 1: Successful login (username hashed, IP anonymized)
log_activity_secure(username="alice_smith", ip_address="192.168.1.100", event_description="Successful Login")

# Example 2: Failed login attempt (username hashed, IP anonymized)
log_activity_secure(username="bob_jones", ip_address="203.0.113.45", event_description="Failed Login Attempt")

# Example 3: User interaction (e.g., viewing a product)
log_activity_secure(username="charlie_brown", ip_address="10.0.0.23", event_description="Viewed Product ID: 12345")

# Example 4: Admin action (might justify logging full username, if policy allows)
# Here, we set minimal_logging to False to demonstrate logging the actual username (e.g., for audit trails)
# In a real system, this would be controlled by strict roles and policies.
print("Logging an admin action with full username (if policy allows, minimal_logging=False).")
log_activity_secure(username="admin_user", ip_address="172.16.0.1", event_description="Modified System Settings", minimal_logging=False)


print("Secure activity logging simulation complete. Please check 'user_activity_secure.log'.")
```
