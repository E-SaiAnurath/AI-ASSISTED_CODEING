#Question 1: Zero-Shot Prompting (Palindrome Number Program)
Write a zero-shot prompt (without providing any examples) to generate
a Python function that checks whether a given number is a palindrome.

Task:

• Record the AI-generated code.

• Test the code with multiple inputs.

• Identify any logical errors or missing edge-case handling.


```python
def is_palindrome_number(n: int) -> bool:
    """
    Checks if a given integer is a palindrome number.

    A palindrome number reads the same forwards and backwards.
    This function handles negative numbers and single-digit numbers.
    It does not use string reversal methods.

    Args:
        n: The integer to check.

    Returns:
        True if the integer is a palindrome, False otherwise.
    """
    # Edge case 1: Negative numbers are not considered palindromes
    # (unless defined differently, but standard definition implies non-negative).
    if n < 0:
        return False

    # Edge case 2: Single-digit numbers (0-9) are always palindromes.
    if 0 <= n < 10:
        return True

    original_num = n
    reversed_num = 0

    # Reverse the number mathematically
    while n > 0:
        # Get the last digit of the number
        digit = n % 10

        # Append the digit to the reversed number
        reversed_num = reversed_num * 10 + digit

        # Remove the last digit from the original number
        n //= 10

    # Compare the original number with its reversed version
    return original_num == reversed_num

# --- Test cases ---
print(f"121 is a palindrome: {is_palindrome_number(121)}")       # Expected: True
print(f"-121 is a palindrome: {is_palindrome_number(-121)}")      # Expected: False
print(f"10 is a palindrome: {is_palindrome_number(10)}")         # Expected: False
print(f"0 is a palindrome: {is_palindrome_number(0)}")           # Expected: True
print(f"7 is a palindrome: {is_palindrome_number(7)}")           # Expected: True
print(f"12321 is a palindrome: {is_palindrome_number(12321)}")   # Expected: True
print(f"12345 is a palindrome: {is_palindrome_number(12345)}")   # Expected: False
print(f"1001 is a palindrome: {is_palindrome_number(1001)}")     # Expected: True
```

```python
def calculate_factorial(n: int) -> int:
    """
    Calculates the factorial of a given non-negative integer.

    The factorial of a non-negative integer n is the product of all positive
    integers less than or equal to n. The factorial of 0 is 1.

    Args:
        n: The non-negative integer to calculate the factorial of.

    Returns:
        The factorial value of n.
    """
    # Edge case: Factorial of 0 is 1
    if n == 0:
        return 1
    # Edge case: Factorial of 1 is 1
    elif n == 1:
        return 1
    # Handle negative input, as factorial is undefined for negative numbers
    elif n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    # Calculate factorial for n > 1
    else:
        factorial_result = 1
        for i in range(1, n + 1):
            factorial_result *= i
        return factorial_result

# --- Test cases ---
print(f"Factorial of 5: {calculate_factorial(5)}")   # Expected: 120
print(f"Factorial of 0: {calculate_factorial(0)}")   # Expected: 1
print(f"Factorial of 1: {calculate_factorial(1)}")   # Expected: 1
print(f"Factorial of 3: {calculate_factorial(3)}")   # Expected: 6

# Test case for negative input
try:
    calculate_factorial(-1)
except ValueError as e:
    print(f"Error for negative input: {e}")
```

#Question 3: Few-Shot Prompting (Armstrong Number Check)
Write a few-shot prompt by providing multiple input-output examples
to guide the AI in generating a Python function to check whether a

• Input: 153 → Output: Armstrong Number

• Input: 370 → Output: Armstrong Number

• Input: 123 → Output: Not an Armstrong Number

```python
# Few-shot examples:
# Input: 153 -> Output: Armstrong Number
# Input: 370 -> Output: Armstrong Number
# Input: 123 -> Output: Not an Armstrong Number

def is_armstrong_number(n: int) -> str:
    """
    Checks if a given integer is an Armstrong number.

    An Armstrong number is a number that is equal to the sum of its own digits
    each raised to the power of the number of digits.

    Args:
        n: The integer to check.

    Returns:
        An appropriate message indicating whether it is an Armstrong number or not.
    """
    # Handle invalid input (non-positive integers)
    if not isinstance(n, int) or n < 0:
        return "Invalid input: Please provide a non-negative integer."

    # Single-digit numbers are Armstrong numbers by definition (e.g., 5 = 5^1)
    if 0 <= n <= 9:
        return f"{n} is an Armstrong Number"

    original_num = n
    num_str = str(n)
    num_digits = len(num_str)
    sum_of_powers = 0

    # Calculate the sum of digits raised to the power of num_digits
    for digit_char in num_str:
        digit = int(digit_char)
        sum_of_powers += digit ** num_digits

    # Compare the sum with the original number
    if sum_of_powers == original_num:
        return f"{original_num} is an Armstrong Number"
    else:
        return f"{original_num} is Not an Armstrong Number"

# --- Test cases ---
print(is_armstrong_number(153)) # Expected: 153 is an Armstrong Number
print(is_armstrong_number(370)) # Expected: 370 is an Armstrong Number
print(is_armstrong_number(123)) # Expected: 123 is Not an Armstrong Number
print(is_armstrong_number(9))   # Expected: 9 is an Armstrong Number
print(is_armstrong_number(0))   # Expected: 0 is an Armstrong Number
print(is_armstrong_number(371)) # Expected: 371 is an Armstrong Number
print(is_armstrong_number(1634))# Expected: 1634 is an Armstrong Number
print(is_armstrong_number(-5))  # Expected: Invalid input: Please provide a non-negative integer.
```

```python
import math

def classify_number(num: int) -> str:
    """
    Classifies an integer as 'Prime', 'Composite', or 'Neither Prime nor Composite'.

    Handles input validation, numbers <= 1, and uses an optimized
    approach for prime checking.

    Args:
        num: The integer to classify.

    Returns:
        A string indicating the classification of the number.
    """
    # 1. Input Validation
    if not isinstance(num, int):
        return "Invalid input: Please provide an integer."
    if num < 0:
        return "Invalid input: Please provide a non-negative integer."

    # 3. Classification for numbers <= 1
    if num <= 1:
        return f"{num} is Neither Prime nor Composite."

    # 4. Optimized Prime/Composite Checking
    # 2 is the only even prime number
    if num == 2:
        return f"{num} is Prime."
    # All other even numbers are composite
    if num % 2 == 0:
        return f"{num} is Composite."

    # Check for odd divisors up to the square root of the number
    # Optimization: Only check odd numbers for divisibility
    # Starting from 3, up to sqrt(num), incrementing by 2
    limit = int(math.sqrt(num)) + 1
    for i in range(3, limit, 2):
        if num % i == 0:
            return f"{num} is Composite."

    # If no divisors found, the number is Prime
    return f"{num} is Prime."

# --- Test cases for demonstration and validation ---
print("--- Testing Valid Inputs ---")
print(classify_number(7))    # Expected: 7 is Prime.
print(classify_number(10))   # Expected: 10 is Composite.
print(classify_number(1))    # Expected: 1 is Neither Prime nor Composite.
print(classify_number(0))    # Expected: 0 is Neither Prime nor Composite.
print(classify_number(2))    # Expected: 2 is Prime.
print(classify_number(3))    # Expected: 3 is Prime.
print(classify_number(97))   # Expected: 97 is Prime.
print(classify_number(99))   # Expected: 99 is Composite.
print(classify_number(169))  # Expected: 169 is Composite. (13*13)

print("\n--- Testing Invalid Inputs ---")
print(classify_number(-5))   # Expected: Invalid input: Please provide a non-negative integer.
print(classify_number(3.14)) # Expected: Invalid input: Please provide an integer.
print(classify_number("abc"))# Expected: Invalid input: Please provide an integer.
```

#Question 5: Zero-Shot Prompting (Perfect Number Check)
Write a zero-shot prompt (without providing any examples) to
generate a Python function that checks whether a given number is a
perfect number.
Task:
• Record the AI-generated code.
• Test the program with multiple inputs.
• Identify any missing conditions or inefficiencies in the logic.

```python
def is_perfect_number(n: int) -> bool:
    """
    Checks if a given positive integer is a perfect number.

    A perfect number is a positive integer that is equal to the sum of its proper positive divisors
    (that is, the sum of its positive divisors excluding the number itself).

    Args:
        n: The integer to check.

    Returns:
        True if the number is a perfect number, False otherwise.
    """
    # Handle edge cases: zero, one, and negative numbers
    if not isinstance(n, int):
        print(f"Invalid input: '{n}' is not an integer.")
        return False # Or raise a TypeError
    if n <= 0:
        # Perfect numbers are defined only for positive integers.
        print(f"Invalid input: Perfect numbers are positive integers. '{n}' is not.")
        return False

    # Initialize sum of proper divisors
    sum_divisors = 0

    # Calculate the sum of proper divisors (excluding the number itself)
    # Avoid using built-in or external libraries for divisor calculation
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i

    # Check if the sum of divisors equals the number itself
    return sum_divisors == n

# --- Test cases ---
print(f"6 is a perfect number: {is_perfect_number(6)}")     # Expected: True (1 + 2 + 3 = 6)
print(f"28 is a perfect number: {is_perfect_number(28)}")   # Expected: True (1 + 2 + 4 + 7 + 14 = 28)
print(f"12 is a perfect number: {is_perfect_number(12)}")   # Expected: False (1 + 2 + 3 + 4 + 6 = 16 != 12)
print(f"1 is a perfect number: {is_perfect_number(1)}")     # Expected: False (edge case: 1 has no proper divisors whose sum equals 1)
print(f"0 is a perfect number: {is_perfect_number(0)}")     # Expected: False (edge case: not a positive integer)
print(f"-6 is a perfect number: {is_perfect_number(-6)}")   # Expected: False (edge case: not a positive integer)
print(f"496 is a perfect number: {is_perfect_number(496)}") # Expected: True
print(f"8128 is a perfect number: {is_perfect_number(8128)}") # Expected: True
print(f"Perfect numbers are positive integers. '0' is not.: {is_perfect_number(0)}")
```

```python
# Few-shot examples:
# Input: 8 -> Output: Even
# Input: 15 -> Output: Odd
# Input: 0 -> Output: Even

def check_even_odd(num_input) -> str:
    """
    Determines if a given number is 'Even', 'Odd', or an 'Invalid Input'.

    Handles input validation, including non-integer values, and correctly
    classifies negative integers.

    Args:
        num_input: The value to check.

    Returns:
        A string indicating the classification ('Even', 'Odd', or error message).
    """
    # Requirement 2: Validate the input and handle non-integer values gracefully.
    try:
        num = int(num_input) # Attempt to convert input to an integer
    except ValueError:
        return f"Invalid input: '{num_input}' is not a valid integer."

    # Requirement 3: Correctly classify negative integers as even or odd.
    # The modulo operator (%) works correctly for negative numbers in Python
    # e.g., -8 % 2 = 0, -15 % 2 = 1 or -1 (depending on Python version/implementation)
    # To ensure consistency and standard even/odd definition, we check absolute value
    # or simply rely on Python's behavior where (negative % 2) will be 0 for even,
    # and 1 or -1 for odd (for odd it's 1 if num_input is positive, -1 if negative)
    # A number is even if it's perfectly divisible by 2, meaning the remainder is 0.
    if num % 2 == 0:
        return f"{num} is Even"
    else:
        return f"{num} is Odd"

# --- Test cases ---
print(check_even_odd(8))       # Expected: 8 is Even
print(check_even_odd(15))      # Expected: 15 is Odd
print(check_even_odd(0))       # Expected: 0 is Even
print(check_even_odd(-4))      # Expected: -4 is Even (Requirement 3)
print(check_even_odd(-7))      # Expected: -7 is Odd (Requirement 3)
print(check_even_odd(3.5))     # Expected: Invalid input (Requirement 2)
print(check_even_odd("hello")) # Expected: Invalid input (Requirement 2)
print(check_even_odd("10"))    # Expected: 10 is Even (handles string integers)
```
